```python
from rdkit.Chem import SaltRemover as sr
remover = sr.SaltRemover()
```

```python
mols = []
for i in range(len(df2['smiles'])):
    mol_i = Chem.MolFromSmiles(df2['smiles'][i])
    mol_i = remover.StripMol(mol_i, dontRemoveEverything=True)
    mols.append(mol_i)
len(mols)
```

```python
with Chem.SDWriter('760mols.sdf') as f:
    for mol in mols:
        f.write(mol)
```

```python
#Calculating Descriptors
i = 0
descriptors = []
for mol in mols:
    descriptors.append(Descriptors.CalcMolDescriptors(mols[i]).values())
    i += 1
len(descriptors)
```

```python
df10 = pd.DataFrame(descriptors, columns=Descriptors.CalcMolDescriptors(mols[0]).keys())
```

```python
df10
```

```python
df10.to_csv('all_descriptors.csv', index=False)
```

```python
#RDKit fingerprint
fpgen1 = AllChem.GetRDKitFPGenerator(fpSize=2048, maxPath=7)
i = 0
l1 = np.zeros((760, 2048), dtype='uint8')
for mol in mols:
    fp1 = fpgen1.GetFingerprintAsNumPy(mols[i])
    l1[i, :] = fp1
    i += 1
rdfp1 = l1
rdfp1.shape 
rdfp1.sum()
```

```python
#Morgan Fingerprints:
fpgen2 = AllChem.GetMorganGenerator()
i = 0
l2 = np.zeros((1, 2048), dtype='uint8')
for mol in mols:
    fp2 = fpgen2.GetFingerprintAsNumPy(mols[i])
    l2 = np.vstack((l2, fp2))
    i += 1
mfp = l2[1:, :]
mfp.shape
mfp.sum()
```
