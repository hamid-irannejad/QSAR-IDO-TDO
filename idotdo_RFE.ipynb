```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report
from sklearn.svm import SVC
from sklearn.svm import LinearSVC
from sklearn.svm import SVR
from sklearn.feature_selection import RFE
from sklearn.feature_selection import RFECV
from sklearn.metrics import roc_curve, RocCurveDisplay, auc, roc_auc_score
from sklearn.preprocessing import LabelEncoder
```

```python
df1 = pd.read_csv('final_descriptors.csv')
```

```python
df2 = pd.read_csv('target_columns.csv')
```

```python
for i in range(760):
    if df2['ido_ic50'].values[i] >= 6.15:
        df2['ido_ic50'].values[i] = 1
    else:
        df2['ido_ic50'].values[i] = 0
```

```python
for i in range(760):
    if df2['tdo_ic50'].values[i] >= 6.00:
        df2['tdo_ic50'].values[i] = 1
    else:
        df2['tdo_ic50'].values[i] = 0
```

```python
df2['tdo_ic50'] = df2['tdo_ic50'].astype('category')
df2['ido_ic50'] = df2['ido_ic50'].astype('category')
```

```python
df = pd.concat([df1, df2], axis=1)
```

```python
df.dropna(axis=0, inplace=True, ignore_index=True)
```

```python
df.loc[:, ['ido_ic50', 'tdo_ic50']].to_csv('target2_columns.csv', index=False)
```

```python
pd.read_csv('target2_columns.csv')
```

```python
y1 = df['ido_ic50']
y2 = df['tdo_ic50']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
x = df.drop(['ido_ic50', 'tdo_ic50'], axis=1)
```

```python
x_scaled = StandardScaler().fit_transform(x)
x_scaled = pd.DataFrame(x_scaled, columns=x.columns)
```

```python
estimator = LinearSVC(dual=False)
```

```python
rfe = RFE(estimator, n_features_to_select=50, step=1)
```

```python
####################################   IDO
rfe1 = rfe.fit(x_scaled, y1)
```

```python
ido_des_rfe = pd.DataFrame(x, columns=x.columns[rfe1.support_])
ido_des_rfe.to_csv('ido_des_rfe.csv', index_label=False)
```

```python
###################################################  TDO
rfe2 = rfe.fit(x_scaled, y2)
```

```python
tdo_des_rfe = pd.DataFrame(x, columns=x.columns[rfe2.support_])
tdo_des_rfe.to_csv('tdo_des_rfe.csv', index_label=False)
```

```python
pd.read_csv('tdo_des_rfe.csv')
```
